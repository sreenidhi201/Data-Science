```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
mydata = pd.read_csv("Mall_Customers.csv")

# Select features for clustering
x = mydata.iloc[:, [3, 4]].values

# Apply K-Means Clustering with 4 clusters
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
km.fit(x)

# Get cluster predictions
pred = km.predict(x)

# Visualize Clusters
# Method 1: Plot clusters with different colors for each cluster
plt.scatter(x[pred == 0, 0], x[pred == 0, 1], c='blue', label='cluster 1')
plt.scatter(x[pred == 1, 0], x[pred == 1, 1], c='red', label='cluster 2')
plt.scatter(x[pred == 2, 0], x[pred == 2, 1], c='green', label='cluster 3')
plt.scatter(x[pred == 3, 0], x[pred == 3, 1], c='yellow', label='cluster 4')
plt.legend()
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Method 2: Alternative visualization method using predicted cluster labels for color-coding
plt.scatter(x[:, 0], x[:, 1], c=pred, cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
